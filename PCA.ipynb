```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from dataLoad import PulsarData
from sklearn import decomposition
from yellowbrick.style import set_palette
```

```python
# For pretty plotting
plt.style.use('seaborn-paper')
set_palette('flatui')
plt.rcParams["font.family"] = "serif"
```

```python
raw_features = PulsarData('HTRU_2').features
raw_targets = PulsarData('HTRU_2').targets
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(raw_features)
pca_features = pca.transform(raw_features)
print(pca.explained_variance_ratio_)
```

```python
plt.figure(1)
x = pca_features[:,0]
y = pca_features[:,1]
_ = sns.scatterplot(x, y, hue=raw_targets, palette='mako')
plt.legend(prop={'size': 8})
plt.title('2 component PCA')
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')
plt.savefig('plots/PCA.pdf')
```
